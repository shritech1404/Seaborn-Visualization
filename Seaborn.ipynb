# Seaborn
* jointplot
* pairplot
* distplot
* countplot
* boxplot
* barplot
* violinplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Jointplot

```python
sns.jointplot('sepal_length', 'petal_length', data=df, kind = 'hex')
```

```python
sns.jointplot('sepal_length', 'petal_length', data=df, kind = 'reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, 'species')
```

## Distplot

```python
sns.distplot(df['sepal_length'])
```

```python
sns.distplot(df['petal_length'])
```

```python
sns.distplot(df['petal_length'],kde=False,bins=10)
```

## Countplot

```python
sns.countplot('species', data=df)
```

```python
sns.countplot(y='species', data=df)
```

## Barplot

```python
sns.barplot('species', 'sepal_width', data=df)
```

```python
sns.barplot(y='species', x='sepal_width', data=df)
```

## Boxplot

```python
sns.boxplot('species', 'sepal_length', data=df)
```

```python
sns.boxplot(y='species', x='petal_width', data=df)
```

```python
sns.boxplot(y='species', x='petal_width', hue='species',data=df)
```

## Violinplot

```python
sns.violinplot(y='species', x='petal_width', data=df)
```

```python
sns.violinplot('species', 'petal_length', data=df)
```
